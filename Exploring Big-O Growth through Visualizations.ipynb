```python
import matplotlib.pyplot as plt
import numpy as np

# Define the functions
def f(n):
    return (0.5 * n * (n - 1) )+ 10

def g(n):
    return n**2

# Generate data for n
n_small = np.arange(1, 11)
n_large = np.arange(1, 10**6 + 1, 10**4)  # Use a step size to keep the plot manageable

# Compute function values
f_small = f(n_small)
g_small = g(n_small)

f_large = f(n_large)
g_large = g(n_large)

# Plot 1: n from 1 to 10
plt.figure(figsize=(12, 6))
plt.plot(n_small, f_small, label='f(n)', color='blue')
plt.plot(n_small, g_small, label='g(n)', color='red')
plt.xlabel('n')
plt.ylabel('Function Value')
plt.title('Functions f(n) and g(n) with n from 1 to 10')
plt.legend()
plt.grid(True)
plt.savefig('plot_1_small_range.png')
plt.show()

# Plot 2: n from 1 to 10^6
plt.figure(figsize=(12, 6))
plt.plot(n_large, f_large, label='f(n)', color='blue')
plt.plot(n_large, g_large, label='g(n)', color='red')
plt.xlabel('n')
plt.ylabel('Function Value')
plt.title('Functions f(n) and g(n) with n from 1 to 10^6')
plt.legend()
plt.grid(True)
plt.savefig('plot_2_large_range.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the functions
def f(n):
    return (0.5 * n * (n - 1) )+ 10

def g(n):
    return n**2

# Generate data for n
n_small = np.arange(1, 11)
n_large = np.arange(1, 10**6 + 1, 10**4)  # Use a step size to keep the plot manageable

# Compute ratio f(n) / g(n)
ratio_small = f(n_small) / g(n_small)
ratio_large = f(n_large) / g(n_large)

# Plot 1: n from 1 to 10
plt.figure(figsize=(12, 6))
plt.plot(n_small, ratio_small, label='f(n) / g(n)', color='green')
plt.xlabel('n')
plt.ylabel('f(n) / g(n)')
plt.title('Ratio of f(n) to g(n) with n from 1 to 10')
plt.legend()
plt.grid(True)
plt.savefig('ratio_plot_1_small_range.png')
plt.show()

# Plot 2: n from 1 to 10^6
plt.figure(figsize=(12, 6))
plt.plot(n_large, ratio_large, label='f(n) / g(n)', color='green')
plt.xlabel('n')
plt.ylabel('f(n) / g(n)')
plt.title('Ratio of f(n) to g(n) with n from 1 to 10^6')
plt.legend()
plt.grid(True)
plt.savefig('ratio_plot_2_large_range.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the new functions
def f_new(n):
    # Ensure the input to sqrt is non-negative
    return np.sqrt(np.maximum(n**5 + 3*n + 1, 0))

def g_new(n):
    return 5 * n**2

# Generate data for n
n_small = np.arange(1, 11)
n_large = np.arange(1, 10**7 + 1, 10**5)  # Use a larger step size for very large n

# Compute function values
f_small_new = f_new(n_small)
g_small_new = g_new(n_small)

f_large_new = f_new(n_large)
g_large_new = g_new(n_large)

# Compute ratios
ratio_small_new = f_small_new / g_small_new
ratio_large_new = f_large_new / g_large_new

# Create a figure with a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Plot 1: New functions with n from 1 to 10
axs[0, 0].plot(n_small, f_small_new, label='f(n) = sqrt(n^5 + 3n + 1)', color='blue')
axs[0, 0].plot(n_small, g_small_new, label='g(n) = 5n^2', color='red')
axs[0, 0].set_xlabel('n')
axs[0, 0].set_ylabel('Function Value')
axs[0, 0].set_title('New Functions f(n) and g(n) (n from 1 to 10)')
axs[0, 0].legend()
axs[0, 0].grid(True)

# Plot 2: New functions with n from 1 to 10^7
axs[0, 1].plot(n_large, f_large_new, label='f(n) = sqrt(n^5 + 3n + 1)', color='blue')
axs[0, 1].plot(n_large, g_large_new, label='g(n) = 5n^2', color='red')
axs[0, 1].set_xlabel('n')
axs[0, 1].set_ylabel('Function Value')
axs[0, 1].set_title('New Functions f(n) and g(n) (n from 1 to 10^7)')
axs[0, 1].legend()
axs[0, 1].grid(True)

# Plot 3: Ratio of new functions with n from 1 to 10
axs[1, 0].plot(n_small, ratio_small_new, label='f(n) / g(n)', color='green')
axs[1, 0].set_xlabel('n')
axs[1, 0].set_ylabel('f(n) / g(n)')
axs[1, 0].set_title('Ratio of f(n) to g(n) (n from 1 to 10)')
axs[1, 0].legend()
axs[1, 0].grid(True)

# Plot 4: Ratio of new functions with n from 1 to 10^7
axs[1, 1].plot(n_large, ratio_large_new, label='f(n) / g(n)', color='green')
axs[1, 1].set_xlabel('n')
axs[1, 1].set_ylabel('f(n) / g(n)')
axs[1, 1].set_title('Ratio of f(n) to g(n) (n from 1 to 10^7)')
axs[1, 1].legend()
axs[1, 1].grid(True)

# Adjust layout for better spacing
plt.tight_layout()

# Save and show the figure with all plots
plt.savefig('combined_plots.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the functions
def f_slow(n):
    return np.log(n)

def g_slow(n):
    return np.sqrt(n)

# Generate data for n
n_small = np.arange(2, 11)  # Start from 2 to avoid divide-by-zero errors in logarithm
n_large = np.arange(2, 10**7 + 1, 10**5)  # Use a larger step size for very large n

# Compute function values
f_small_slow = f_slow(n_small)
g_small_slow = g_slow(n_small)

f_large_slow = f_slow(n_large)
g_large_slow = g_slow(n_large)

# Compute ratio f(n) / g(n)
ratio_small_slow = f_small_slow / g_small_slow
ratio_large_slow = f_large_slow / g_large_slow

# Create a figure with a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Plot 1: Functions with n from 2 to 10
axs[0, 0].plot(n_small, f_small_slow, label='f(n) = log(n)', color='blue')
axs[0, 0].plot(n_small, g_small_slow, label='g(n) = sqrt(n)', color='red')
axs[0, 0].set_xlabel('n')
axs[0, 0].set_ylabel('Function Value')
axs[0, 0].set_title('Functions f(n) and g(n) (n from 2 to 10)')
axs[0, 0].legend()
axs[0, 0].grid(True)

# Plot 2: Functions with n from 2 to 10^7
axs[0, 1].plot(n_large, f_large_slow, label='f(n) = log(n)', color='blue')
axs[0, 1].plot(n_large, g_large_slow, label='g(n) = sqrt(n)', color='red')
axs[0, 1].set_xlabel('n')
axs[0, 1].set_ylabel('Function Value')
axs[0, 1].set_title('Functions f(n) and g(n) (n from 2 to 10^7)')
axs[0, 1].legend()
axs[0, 1].grid(True)

# Plot 3: Ratio of functions with n from 2 to 10
axs[1, 0].plot(n_small, ratio_small_slow, label='f(n) / g(n)', color='green')
axs[1, 0].set_xlabel('n')
axs[1, 0].set_ylabel('f(n) / g(n)')
axs[1, 0].set_title('Ratio of f(n) to g(n) (n from 2 to 10)')
axs[1, 0].legend()
axs[1, 0].grid(True)

# Plot 4: Ratio of functions with n from 2 to 10^7
axs[1, 1].plot(n_large, ratio_large_slow, label='f(n) / g(n)', color='green')
axs[1, 1].set_xlabel('n')
axs[1, 1].set_ylabel('f(n) / g(n)')
axs[1, 1].set_title('Ratio of f(n) to g(n) (n from 2 to 10^7)')
axs[1, 1].legend()
axs[1, 1].grid(True)

# Adjust layout for better spacing
plt.tight_layout()

# Save and show the combined figure with all plots
plt.savefig('combined_plots.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the new functions
def f_log2(n):
    return np.log2(n)

def g_log10(n):
    return np.log10(n)

# Generate data for n
n_small = np.arange(2, 11)  # Start from 2 to avoid divide-by-zero errors in logarithm
n_large = np.arange(2, 10**7 + 1, 10**5)  # Use a larger step size for very large n

# Compute function values
f_small_log2 = f_log2(n_small)
g_small_log10 = g_log10(n_small)

f_large_log2 = f_log2(n_large)
g_large_log10 = g_log10(n_large)

# Compute ratio f(n) / g(n)
ratio_small_log = f_small_log2 / g_small_log10
ratio_large_log = f_large_log2 / g_large_log10

# Create a figure with subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 10), sharex='col', sharey='row')

# Plot 1: Logarithmic functions with n from 2 to 10
axs[0, 0].plot(n_small, f_small_log2, label='f(n) = log2(n)', color='blue')
axs[0, 0].plot(n_small, g_small_log10, label='g(n) = log10(n)', color='red')
axs[0, 0].set_xlabel('n')
axs[0, 0].set_ylabel('Function Value')
axs[0, 0].set_title('Logarithmic Functions f(n) and g(n) with n from 2 to 10')
axs[0, 0].legend()
axs[0, 0].grid(True)

# Plot 2: Logarithmic functions with n from 2 to 10^7
axs[0, 1].plot(n_large, f_large_log2, label='f(n) = log2(n)', color='blue')
axs[0, 1].plot(n_large, g_large_log10, label='g(n) = log10(n)', color='red')
axs[0, 1].set_xlabel('n')
axs[0, 1].set_ylabel('Function Value')
axs[0, 1].set_title('Logarithmic Functions f(n) and g(n) with n from 2 to 10^7')
axs[0, 1].legend()
axs[0, 1].grid(True)

# Plot 3: Ratio of new functions with n from 2 to 10
axs[1, 0].plot(n_small, ratio_small_log, label='f(n) / g(n)', color='green')
axs[1, 0].set_xlabel('n')
axs[1, 0].set_ylabel('f(n) / g(n)')
axs[1, 0].set_title('Ratio of f(n) to g(n) with n from 2 to 10')
axs[1, 0].legend()
axs[1, 0].grid(True)

# Plot 4: Ratio of new functions with n from 2 to 10^7
axs[1, 1].plot(n_large, ratio_large_log, label='f(n) / g(n)', color='green')
axs[1, 1].set_xlabel('n')
axs[1, 1].set_ylabel('f(n) / g(n)')
axs[1, 1].set_title('Ratio of f(n) to g(n) with n from 2 to 10^7')
axs[1, 1].legend()
axs[1, 1].grid(True)

# Adjust layout to prevent overlap
plt.tight_layout()

# Save and show the combined plot
plt.savefig('combined_log_plots.png')
plt.show()
```
